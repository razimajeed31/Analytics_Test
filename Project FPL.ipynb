```python
import requests
import pandas as pd
from pandas.io.json import json_normalize
```

```python
pd.set_option('display.max_columns',None)
```

```python
response = requests.get("https://fantasy.premierleague.com/api/bootstrap-static/")
players = response.json()['elements']
player_database = pd.DataFrame(players)
```

```python
player_data = player_database[['id','first_name','second_name','element_type','form','now_cost',
                         'points_per_game','chance_of_playing_next_round','status','team','team_code','total_points',
                        'value_form','minutes','goals_scored','assists','clean_sheets','goals_conceded','own_goals',
                         'penalties_saved','penalties_missed','yellow_cards','red_cards','saves','bonus','bps',
                        'influence','creativity','threat','ict_index','influence_rank','influence_rank_type',
                        'creativity_rank','creativity_rank_type','threat_rank','threat_rank_type','ict_index_rank',
                        'ict_index_rank_type','corners_and_indirect_freekicks_order','direct_freekicks_order',
                         'penalties_order','event_points']]
```

```python
player_data[['form','points_per_game','value_form','influence',
             'creativity','threat','ict_index']] = player_data[[
    'form','points_per_game','value_form','influence','creativity','threat','ict_index']].astype(float)
```

```python
player_data
```

```python
#Player Database split
forwards = player_data[player_data['element_type'].isin([4])]
midfielders = player_data[player_data['element_type'].isin([3])]
defenders = player_data[player_data['element_type'].isin([2])]
goalkeepers = player_data[player_data['element_type'].isin([1])]
```
